## Exercise 1: Calculate the multiplication and sum of two numbers

```python
# Given two integer numbers, write a Python program to return their product only if 
# the product is equal to or lower than 1000. Otherwise, return their sum.
```

```python
num1 = 40
num2 = 30

sum = num1 + num2
```

```python
if sum <= 1000:
    print(num1 * num2)
else:
    print(sum)
```

```python
def cal(num1,num2):
    product = num1 * num2
    if(product <= 1000):
        return product
    else:
        return num1 + num2
```

```python
result = cal(20,30)
print(result)
```

```python
result = cal(30,40)
print(result)
```
