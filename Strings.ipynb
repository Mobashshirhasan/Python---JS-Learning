# Negative String Slicing

```python
name = "Mobashshir"

print(name[0:3])
```

```python
name.
```

```python
# print(name[-4: -1])
print(name[-4: -1])  #your slice is from -1 to -4 â†’ and since Python slicing works left to right, this means the start index (-1) is after the *end index (-4)**.
# convert your negative indices into corresponding positive indices 
print(name[1: 4])
```

```python
# If you want the last 3 characters in reverse, you can do:
# string[start : stop : step]

print(name[-1:-4:-1])
```

```python
print(name[:4])     # 0, 1, 2, 3
```

```python
print(name[:4]) # length -1
```

```python
print(name[1:])     #is same as [1:length-1]
```
